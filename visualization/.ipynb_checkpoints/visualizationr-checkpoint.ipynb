```python
# labels and figsize
```

```python
area = np.array([700,800,1000,1500,2000])
price = np.array([40,45,48,60,70])
```

```python
plt.figure()

plt.title("Area vs Price Plot")
plt.xlabel("Area")
plt.ylabel("Price")

plt.plot(area,price)

plt.show()
```

```python
#xlim and ylim
```

```python
area = np.array([700,800,1000,1500,2000])
price = np.array([40,45,48,60,70])

plt.figure()

plt.title("Area vs Price Plot")
plt.xlabel("Area")
plt.ylabel("Price")

plt.xlim(400,2500)
plt.ylim(10,100)

plt.plot(area,price)

plt.show()
```

```python
#xticks and yticks
```

```python
area = np.array([700,800,1000,1500,2000])
price = np.array([40,45,48,60,70])

plt.figure()

plt.title("Area vs Price Plot")
plt.xlabel("Area")
plt.ylabel("Price")

plt.xticks(rotation=45)
plt.yticks(rotation=90)

plt.plot(area,price)

plt.show()
```

```python
# Multiple plots
```

```python
area = np.array([700,800,1000,1500,2000])
price_mumbai = np.array([40,45,48,60,70])
price_thane = np.array([35,41,42,58,65])

plt.figure(figsize=(10,6))

plt.title("Area vs Price plot")
plt.xlabel("Area")
plt.ylabel("Price")

plt.plot(area,price_mumbai,color="yellow")
plt.plot(area,price_thane,color="red")

plt.legend(labels=["Mubai","Thane"])

plt.show()
```

# Histogram

<b> Frequency plot for continuous data </b>

```python
price = np.array([35,41,45,47,50,51,51,59,59,60,60,61,65,70,75,76,77,78,79,80])
```

```python
# bins = 4
# (80 - 35)/4 = 11


# bin1 -> 35 - 46 -> 3
# bin2 -> 46 - 57 -> 4
# bin3 -> 57 - 68 -> 6 
# bin4 -> 68 - 79 -> 8 
```

```python
plt.hist(price,bins=4)
plt.show()
```

```python
price = np.array([35,41,45,47,50,51,51,59,59,60,60,61,65,70,75,76,77,78,79,80])
```

```python
# bins = 3
# (80 - 35)/3 = 15

# bin1 -> 35 - 50 - 4
# bin2 -> 50 - 65 - 8
# bin3 -> 65 - 80 - 8
```

```python
plt.hist(price,bins=3)
plt.show()
```

```python
plt.hist(price)
plt.show()
```

```python
# car Dataset
```

```python
cars = pd.read_csv("cars.csv")
```

```python
cars.head()
```

```python
plt.figure()

plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.hist(cars["price"])


plt.show()
```

```python
price_mean = cars["price"].mean()
price_median = cars["price"].median()
```

```python
plt.figure()

plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.axvline(price_mean,color="green")
plt.axvline(price_median,color="red")

plt.hist(cars["price"])


plt.show()
```

```python
plt.figure()

plt.title("Width distribution")
plt.xlabel("Width")
plt.ylabel("Frequency")


plt.hist(cars["width"])


plt.show()
```

```python
# pandas -> numpy + matplotlib
```

```python
cars["width"].hist()
plt.show()
```

# Barplot

<b> frequency plot for catagorical data</b>

```python
place = ["Mumbai","Thane","Kalyan","Panvel"]
no_of_flats = [800,750,900,1000]
```

```python
plt.figure()
plt.bar(place,no_of_flats)
plt.show()
```

```python
plt.figure()
plt.barh(place,no_of_flats)

plt.title("Place vs Flat Frequency")
plt.xlabel("No of flats")
plt.ylabel("Place")

plt.show()
```

```python
# barplot using pandas
```

```python
cars["body-style"].value_counts().plot(kind="bar")
plt.show()
```

```python
cars["body-style"].value_counts().plot(kind="barh")
plt.show()
```

# pie chart
<b>Frequency plot for categorical plot, frequency is representated in the form percentage</b>

```python
plt.figure(figsize=(8,6))
cars["engine-type"].value_counts().plot(kind="pie")
plt.show()
```

```python
plt.figure(figsize=(10,10))
cars["engine-type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(10,10))
cars["engine-type"].value_counts().plot(kind="pie", explode=(0,0,0,0.2,0,0,0))
plt.show()
```

# Subplots

```python
area = np.array([700,800,1000,1500,2000])
price = np.array([40,45,48,60,70])
```

```python
fig,axes = plt.subplots(2,2,figsize=(10,10))


axes[0,0].hist(cars["width"])

axes[0,1].bar(place,no_of_flats)

axes[1,0].scatter(cars["engine-size"],cars["price"])

axes[1,1].plot(area,price)

plt.show()
```
